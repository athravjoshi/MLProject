```python
df2=pd.read_excel('/content/Country-Code.xlsx')
df2.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df2,on='Country Code', how='left')
final_df.head(2)
```

```python
final_df.dtypes
```

```python
country_name=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_name[:3],autopct='%1.3f%%',)
plt.show()
```

Obervation:Zomato maximum trancatation is happning in india followed by UK and US


```python
final_df.columns
```

```python
rating=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
rating
```

Observations



1.   when rating is between 4.5 to 4.9 ---> Excellent
2.   when rating is between 4.0 to 4.4 ---> Very Good
3.   when rating is between 3.9 to 3.5 ---> Good
4.   when rating is between 3.4 to 2.5 ---> average

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating', y='Rating Count', data=rating)
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count',hue='Rating color', data=rating, palette=['silver','red','orange','yellow','green','green'])
```

Observation

1.   Not Rated count is high
2.   most number of rating is between 2.5 to 3.9




```python
# count plot

sns.countplot(x='Rating color', data=rating, palette=['silver','red','orange','yellow','green','green'])
```
